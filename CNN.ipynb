```python
train_dir = '/content/fruits_dataset/fruits-360_100x100/fruits-360/Training'
test_dir = '/content/fruits_dataset/fruits-360_100x100/fruits-360/Test'

from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Rescale pixel values to [0, 1]
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical'  # explicitly categorical since multiple classes
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical'
)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout , Input

img_height, img_width = 100, 100
model = Sequential([
    Input(shape=(img_height, img_width, 3)),
    Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    MaxPool2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPool2D(pool_size=(2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(199, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Train classes:", train_generator.num_classes)
print("Test classes:", test_generator.num_classes)
train_classes = set(train_generator.class_indices.keys())
test_classes = set(test_generator.class_indices.keys())
print("Classes in train but not in test:", train_classes - test_classes)
print("Classes in test but not in train:", test_classes - train_classes)

model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```
